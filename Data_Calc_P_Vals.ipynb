```python
import pandas as pd
from scipy import stats

# 1. Load your formatted data
df = pd.read_csv('clean_data.csv')

# 2. Corrected: Grouping by the Experiment Code Set
experiments = {
    "Experiment 1 (PT Code Set)": ['PT1', 'PT2', 'PT3'],
    "Experiment 2 (ST Code Set)": ['ST1', 'ST2', 'ST3'],
    "Experiment 3 (HT Code Set)": ['HT1', 'HT2', 'HT3']
}

# 3. Loop through and calculate the p-values
for exp_name, labels in experiments.items():
    print(f"\n--- {exp_name} ---")

    # Filter for just the 3 structures within this specific experiment
    df_exp = df[df['Experiment_Type'].isin(labels)]

    # Calculate Time p-value (Alexander-Govern Test for unequal variances)
    g1 = df_exp[df_exp['Experiment_Type'] == labels[0]]['Time']
    g2 = df_exp[df_exp['Experiment_Type'] == labels[1]]['Time']
    g3 = df_exp[df_exp['Experiment_Type'] == labels[2]]['Time']

    ag_res = stats.alexandergovern(g1, g2, g3)
    print(f"Time p-value: {ag_res.pvalue}")

    # Calculate Accuracy p-value (Kruskal-Wallis)
    groups = [group['Accuracy'].values for name, group in df_exp.groupby('Experiment_Type')]
    stat, acc_pval = stats.kruskal(*groups)
    print(f"Accuracy p-value: {acc_pval}")
```
